```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('./drug200.csv', sep=',')
```

```python
print(dataset.head())
```

```python
dummies = pd.get_dummies(dataset.BP)
merged = pd.concat([dataset, dummies], axis='columns')
print(merged)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize encoder
label_encoder = LabelEncoder()

# Convert categorical columns
for col in dataset.select_dtypes(include=["object"]).columns:
    dataset[col] = label_encoder.fit_transform(dataset[col])

print(dataset.head())  
```

```python
df_encoded = pd.get_dummies(dataset, columns=['Sex', 'BP', 'Cholesterol', 'Drug'], drop_first=True)
print(df_encoded.head())
```

```python
dataset.replace({
    'Sex': {'F': 0, 'M': 1},
    'BP': {'LOW': 0, 'NORMAL': 1, 'HIGH': 2},
    'Cholesterol': {'NORMAL': 0, 'HIGH': 1},
    'Drug': {'drugA': 0, 'drugB': 1, 'drugC': 2, 'drugX': 3, 'drugY': 4}
}, inplace=True)

print(dataset.head)
```

```python
print(dataset)
```

```python

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
```

```python
X = dataset.drop(columns=['Drug'])  # Features
y = dataset['Drug']  # Target variable
cv_scores = cross_val_score(clf, X , y, cv = 10)
```

```python

print(f"Cross-validation accuracy: {cv_scores.mean():.4f}")
```

```python
print("Cross validation scores", cv_scores)
print("Mean Accuracy: ", cv_scores.mean())
```

```python
mean_accuracy = cv_scores.mean()
print("Average accuracy: ", mean_accuracy)
print(f"Average Accuracy from 10-fold Cross-Validation: {cv_scores.mean():.4f}")
```

```python
clf_gini = DecisionTreeClassifier(criterion="gini", random_state=0)
cv_gini = cross_val_score(clf_gini, X, y, cv=10)
print(f"Gini Accuracy: {cv_gini.mean():.4f}")
```

```python
clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=0)
cv_entropy = cross_val_score(clf_entropy, X, y, cv=10)
print(f"Entropy Accuracy: {cv_entropy.mean():.4f}")
```

```python
clf_log_loss = DecisionTreeClassifier(criterion="log_loss", random_state=0)
cv_log_loss = cross_val_score(clf_log_loss, X, y, cv=10)
print(f"Log Loss Accuracy: {cv_log_loss.mean():.4f}")
```

```python
clf_default = DecisionTreeClassifier(max_depth=None, random_state=0)
cv_default = cross_val_score(clf_default, X, y, cv=10)
print(f"Default max_depth Accuracy: {cv_default.mean():.4f}")
clf_depth_3 = DecisionTreeClassifier(max_depth=3, random_state=0)
cv_depth_3 = cross_val_score(clf_depth_3, X, y, cv=10)
print(f"Max Depth 3 Accuracy: {cv_depth_3.mean():.4f}")
clf_depth_5 = DecisionTreeClassifier(max_depth=5, random_state=0)
cv_depth_5 = cross_val_score(clf_depth_5, X, y, cv=10)
print(f"Max Depth 5 Accuracy: {cv_depth_5.mean():.4f}")
clf_depth_10 = DecisionTreeClassifier(max_depth=10, random_state=0)
cv_depth_10 = cross_val_score(clf_depth_10, X, y, cv=10)
print(f"Max Depth 10 Accuracy: {cv_depth_10.mean():.4f}")

```
